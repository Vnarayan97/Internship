<B>STATISTICAL ANALYSIS!!<BR>

```python
from scipy import stats
```

```python
reject=[]
nonreject=[]
def check(target_var,feat):
    print(f"\n\nH0- THE mean of {feat.name} is equal for both categories of {target_var.name}")
    print(f"\n\nHa- The mean of {feat.name} is not equal for both categories of {target_var.name}")
    grp1=feat[target_var==0]
    grp2=feat[target_var==1]
    ttest=stats.ttest_ind(grp1,grp2)
    print(f'\n THE RESULT OF THE TTEST IS {ttest}')
    rej=ttest[1]<0.05
    if rej:
        print("\n\nREJECT H0")
        reject.append(col)
        print('\n')
        print('----------------------------------------------------------')
    else:
        print("\n\nNOT REJECT H0")
        nonreject.append(col)
        print('\n')
        print('-----------------------------------------------------------')    
```

```python
for col in columns:
    rej=check(df1['label'],df1[col])
print(f"\nTHE REJECT LIST IS {reject}")
print(f"\nThe not reject list is {nonreject}")
```

<B> DETERMINING THE EIGEN VALUES AND EIGNE VECTORS FOR FURTHER PLOTS!

```python
eig_vals,eig_vecs=np.linalg.eig(cov_matrix)
```

```python
eig_pairs=[(np.abs(eig_vals[i]),eig_vecs[:,i]) for i in range(len(eig_vals))]
```

```python
eig_pairs
```

```python
total=sum(eig_vals)
```

```python
var_exp=[(i/total)*100 for i in sorted(eig_vals,reverse=True)]
```

```python
results=[]
names=[]
models=[]
models.append(("LOGISTICS",lr))
models.append(("KNN",knn))
models.append(("DECISIONTREE",dt))
models.append(("RANDOMFOREST",rfc))
models.append(("ADABOOST",adc))
models.append(("BAGGINGCLASSIFIER",bc))
models.append(("GRADIENTBOOSTING",gbc))
```

```python
for name,model in models:
    kfold=KFold(shuffle=True,random_state=0,n_splits=3)
    cross_score=cross_val_score(model,x_train,y_train,cv=kfold)
    results.append(cross_score)
    names.append(name)
    print(f"{name}---{np.mean(cross_score)}---{(np.var(cross_score,ddof=1))}")
fig=plt.figure(figsize=(8,8))
fig.suptitle("MODELS COMPARISION")
ax=fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names,rotation=45)
plt.show()    
```
