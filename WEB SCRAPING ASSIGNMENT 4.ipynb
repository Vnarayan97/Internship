```python
driver=webdriver.Chrome(r'C:\Users\VNarayan\Downloads\chromedriver.exe')
```

```python
rank=[]
name=[]
artist=[]
ud=[]
views=[]
lis=[]
```

GETTING THE URL

```python
try:
    url='https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
    driver.get(url)
    time.sleep(5)
    tab=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td")
    n=1
    for i in tab:
        if n<=180:
            lis.append(i.text)
            n+=1
        else:
            break
    n=6
    x=[lis[i:i + n] for i in range(0, len(lis), n)]
    for i in x:
        rank.append(i[0])
        name.append(i[1])
        artist.append(i[2])
        views.append(i[3])
        ud.append(i[4])
    df=pd.DataFrame({'RANK':rank,'NAME':name,'ARTIST':artist,'VIEWS':views,'UPLOAD_DATE':ud})
except:
    print("ERROR")
```
